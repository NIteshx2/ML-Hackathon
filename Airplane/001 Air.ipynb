```python
def mean_encodings(df , cols) :
    for col in cols :
        df['{}_enc'.format(col)] = df.groupby(col)['class'].transform('mean')
        df['{}_enc'.format(col)].fillna(0.4453, inplace=True)
    return df

cols = [ 'Accident_Type_Code' , 'Days_Since_Inspection' , 'Violations' , 'Total_Safety_Complaints']

conc = mean_encodings(conc , cols)
```

```python
# train 
train = conc[pd.notnull(conc['class'])]
```

```python
test = conc[pd.isnull(conc['class'])]
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
