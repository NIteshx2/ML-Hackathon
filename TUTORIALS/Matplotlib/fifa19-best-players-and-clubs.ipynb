## Best Clubs ranked :
* Good Clubs have the best average for their starting 11*

```python
teams = {}
#plots = (fast_plot())
for club in df['Club'].unique() :
    pl = int(11)
    overall = df['Overall'][ df['Club'] == club].sort_values( ascending = False).head(pl).sum()
    overall /= (pl*1.0)
    teams[club] = overall
    

```

```python
df_teams = pd.DataFrame.from_dict(teams , orient = 'index' , columns = ['overall'])
df_teams = df_teams.sort_values(by = 'overall' , ascending = 0).head(20)
df_teams['club'] = df_teams.index
plt.figure(1 , figsize = (15 , 6))
plt.barh(y = 'club' , width = 'overall' , data = df_teams )
plt.xlim(70 , 93)
plt.xlabel("Average Player OVR for " + str(pl) + " players")
plt.show()
```

## Clubs ranked by top 16 players 
* Substitutes are needed to win long competitions, when 1st choice players tend to get injured

```python
teams = {}
#plots = (fast_plot())
for club in df['Club'].unique() :
    pl = int(16)
    overall = df['Overall'][ df['Club'] == club].sort_values( ascending = False).head(pl).sum()
    overall /= (pl*1.0)
    teams[club] = overall
    

```

```python
df_teams = pd.DataFrame.from_dict(teams , orient = 'index' , columns = ['overall'])
df_teams = df_teams.sort_values(by = 'overall' , ascending = 0).head(20)
df_teams['club'] = df_teams.index
plt.figure(1 , figsize = (15 , 6))
plt.barh(y = 'club' , width = 'overall' , data = df_teams )
plt.xlim(70 , 93)
plt.xlabel("Average Player OVR for " + str(pl) + " players")
plt.show()
```
