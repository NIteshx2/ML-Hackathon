```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
os.listdir('../input/')
```

```python
import pandas_profiling as pp
```

```python
train = pd.read_csv('../input/Train.csv')
test = pd.read_csv('../input/Test.csv')
sample = pd.read_excel('../input/Sample_Submission.xlsx')
```

```python
sample
```

```python
df = train.append(test)
df.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
from sklearn.preprocessing import OneHotEncoder 
  
# creating one hot encoder object with categorical feature 0 
# indicating the first column 
# onehotencoder = OneHotEncoder(categorical_features = [0])
# data = onehotencoder.fit_transform(data).toarray() 

for col in df.columns :
    if col != 'grade':
        if df[col].nunique() == 2 :
            df[col] = le.fit_transform(df[col])
        elif df[col].nunique() > 2 and df[col].nunique() < 5 :
            dums = pd.get_dummies(df[col])
            df = df.drop([col] , axis = 1)
            df = pd.concat([df , dums] , axis = 1)

df
```

```python
test = df[df['grade'].isna()]
train = df[~df['grade'].isna()]

train
```

### Train test split

```python
# sample
from sklearn.model_selection import train_test_split
# train_test_split(train , test, test_size=0.3, random_state=42)
```

```python
X = train.drop(['grade'] , axis = 1)
y = train['grade']
```

```python
# train_y.value_counts()
```

```python
# train_X , val_X , train_y , val_y = train_test_split(X, y, test_size=0.3, random_state=42)
train_X = X
train_y = y
val_X = test.drop(['grade'] , axis = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import log_loss
# Create the model with 100 trees
model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
# Fit on training data
model.fit(train_X , train_y)
```
